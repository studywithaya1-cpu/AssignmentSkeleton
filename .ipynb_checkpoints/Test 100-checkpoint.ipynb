```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_rows", 120)
pd.set_option("display.max_columns", 200)

try:
 from IPython.display import display
except:
 display = print


dt = 0.04
pitch_length = 105
pitch_width = 68

window_sec = 300
buffer_sec = 30
rolling_sec = 60

max_speed = 10
match_window_sec = 10


game1_home = "Sample_Game_1_RawTrackingData_Home_Team.csv"
game1_away = "Sample_Game_1_RawTrackingData_Away_Team.csv"
game2_home = "Sample_Game_2_RawTrackingData_Home_Team.csv"
game2_away = "Sample_Game_2_RawTrackingData_Away_Team.csv"


def read_and_build_tracking(path):
    raw = pd.read_csv(path, header=2)

    df = raw.iloc[:, :3].copy()
    df.columns = ["Period", "Frame", "Time"]

    columns = list(raw.columns)
    i = 3

    while i < len(columns) - 1:
        x_name = columns[i]
        y_name = columns[i + 1]

        if "Unnamed" not in str(x_name) and "Unnamed" in str(y_name):
            player = str(x_name).replace(" ", "")
            df[player + "_x"] = raw[x_name]
            df[player + "_y"] = raw[y_name]
            i += 2
        else:
            i += 1

    df["Time"] = pd.to_numeric(df["Time"], errors="coerce")
    df["Period"] = pd.to_numeric(df["Period"], errors="coerce")
    df["Frame"] = pd.to_numeric(df["Frame"], errors="coerce")

    players = []
    for col in df.columns:
        if col.startswith("Player") and col.endswith("_x"):
            players.append(col[:-2])

    players = sorted(players)
    return df, players


def to_metric_coordinates(df):
    out = df.copy()

    for col in out.columns:
        if col.endswith("_x"):
            out[col] = (out[col] - 0.5) * pitch_length
        elif col.endswith("_y"):
            out[col] = (out[col] - 0.5) * pitch_width

    return out


def add_match_clock(df):
    out = df.sort_values(["Period", "Frame"]).reset_index(drop=True).copy()
    out["match_time_s"] = np.nan

    periods = sorted(out["Period"].dropna().unique())
    running_end = 0.0

    for p in periods:
        idx = out["Period"] == p
        times = out.loc[idx, "Time"]

        start = times.min()
        end = times.max()

        out.loc[idx, "match_time_s"] = (times - start) + running_end
        running_end += (end - start)

    return out


def detect_substitutions(df, players):
    g = df.sort_values(["match_time_s", "Frame"]).reset_index(drop=True)
    t = g["match_time_s"].values

    exits = []
    enters = []

    for p in players:
        is_on = g[p + "_x"].notna().values.astype(int)
        change = np.diff(is_on)

        exits_idx = np.where(change == -1)[0] + 1
        enters_idx = np.where(change == 1)[0] + 1

        for i in exits_idx:
            exits.append((i, p))
        for i in enters_idx:
            enters.append((i, p))

    exits.sort(key=lambda item: t[item[0]])

    used_enters = set()
    subs = []

    for ex_i, p_out in exits:
        ex_t = t[ex_i]
        best_dt = None
        best_enter = None

        for en_i, p_in in enters:
            if en_i in used_enters:
                continue

            dt_diff = abs(t[en_i] - ex_t)
            if dt_diff <= match_window_sec and (best_dt is None or dt_diff < best_dt):
                best_dt = dt_diff
                best_enter = (en_i, p_in)

        if best_enter is not None:
            en_i, p_in = best_enter
            used_enters.add(en_i)

            sub_time = (ex_t + t[en_i]) / 2
            subs.append({
                "match_time_s": sub_time,
                "minute": sub_time / 60,
                "player_out": p_out,
                "player_in": p_in
            })

    subs = pd.DataFrame(subs)
    if subs.empty:
        return subs

    return subs.sort_values("match_time_s").reset_index(drop=True)


def compute_team_intensity(df_metric, players):
    n = len(df_metric)

    team_speed_sum = np.zeros(n)
    active_count = np.zeros(n)

    for p in players:
        x = df_metric[p + "_x"].to_numpy(float)
        y = df_metric[p + "_y"].to_numpy(float)

        active = (~np.isnan(x)) & (~np.isnan(y))
        moved = active & np.roll(active, 1)
        moved[0] = False

        speed = np.full(n, np.nan)
        idx = np.where(moved)[0]

        dx = x[idx] - x[idx - 1]
        dy = y[idx] - y[idx - 1]

        dist = np.sqrt(dx * dx + dy * dy)
        speed[idx] = np.clip(dist / dt, 0, max_speed)

        active_count += active.astype(float)
        team_speed_sum += np.nan_to_num(speed, nan=0.0)

    intensity_value = np.where(active_count > 0, team_speed_sum / active_count, np.nan)

    intensity = pd.DataFrame({
        "match_time_s": df_metric["match_time_s"].to_numpy(float),
        "minute": df_metric["match_time_s"].to_numpy(float) / 60,
        "intensity": intensity_value,
        "active_players": active_count
    })

    roll_n = max(1, int(round(rolling_sec / dt)))
    intensity["intensity_roll"] = intensity["intensity"].rolling(
        roll_n, center=True, min_periods=1
    ).mean()

    return intensity


def compute_svs(subs, intensity, team_label):
    if subs.empty:
        return pd.DataFrame()

    rows = []
    for _, sub in subs.iterrows():
        t0 = float(sub["match_time_s"])

        before = intensity[
            (intensity["match_time_s"] >= t0 - window_sec) &
            (intensity["match_time_s"] <  t0 - buffer_sec)
        ]

        after = intensity[
            (intensity["match_time_s"] >  t0 + buffer_sec) &
            (intensity["match_time_s"] <= t0 + window_sec)
        ]

        if before.empty or after.empty:
            continue

        b = float(before["intensity_roll"].mean())
        a = float(after["intensity_roll"].mean())

        rows.append({
            "minute": float(sub["minute"]),
            "player_out": sub["player_out"],
            "player_in": sub["player_in"],
            "SVS": a - b,
            "team": team_label
        })

    return pd.DataFrame(rows)


def plot_intensity(intensity, subs, title):
    plt.figure(figsize=(12, 4))
    plt.plot(intensity["minute"], intensity["intensity_roll"])

    if not subs.empty:
        for _, sub in subs.iterrows():
            plt.axvline(float(sub["minute"]), linestyle="--", alpha=0.5)

    plt.title(title)
    plt.xlabel("Match Minute")
    plt.ylabel("Intensity (rolling)")
    plt.tight_layout()
    plt.show()


all_svs = []

paths = {
    "Game 1 Home": game1_home,
    "Game 1 Away": game1_away,
    "Game 2 Home": game2_home,
    "Game 2 Away": game2_away
}

for label, path in paths.items():
    print("Processing:", label)

    df, players = read_and_build_tracking(path)
    print("Players detected:", len(players))

    df = add_match_clock(df)

    subs = detect_substitutions(df, players)
    print("Substitutions found:", len(subs))

    df_metric = to_metric_coordinates(df)

    intensity = compute_team_intensity(df_metric, players)

    svs = compute_svs(subs, intensity, label)
    all_svs.append(svs)

    plot_intensity(intensity, subs, f"{label} - Team Intensity")


if all_svs:
    svs_all = pd.concat(all_svs, ignore_index=True)
else:
    svs_all = pd.DataFrame()

if svs_all.empty:
    print("No SVS results to plot.")
else:
    svs_all["bin"] = (svs_all["minute"] // 10) * 10
    summary = svs_all.groupby("bin")["SVS"].mean()

    plt.figure(figsize=(10, 4))
    plt.bar(summary.index.astype(int), summary.values)
    plt.axhline(0)
    plt.title("Average SVS by Substitution Time")
    plt.ylabel("Mean SVS")
    plt.xlabel("Match minute (10-minute bins)")
    plt.xticks(summary.index.astype(int), rotation=45)
    plt.tight_layout()
    plt.show()

    display(svs_all.head(10))
```
